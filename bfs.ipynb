```python
import heapq

# Define the graph as an adjacency list with heuristic values for each node
graph = {
    'A': [('B', 1), ('C', 3)],
    'B': [('D', 2), ('E', 4)],
    'C': [('F', 5)],
    'D': [],
    'E': [('G', 1)],
    'F': [],
    'G': []
}

# Define heuristic values for each node (lower values indicate closer to goal)
heuristic = {
    'A': 7,
    'B': 6,
    'C': 5,
    'D': 4,
    'E': 3,
    'F': 6,
    'G': 0   # Goal node has a heuristic of 0
}

def best_first_search(start, goal):
    # Priority queue for BFS with heuristic priority
    open_list = []
    heapq.heappush(open_list, (heuristic[start], start))  # (heuristic, node)
    
    # Track visited nodes
    visited = set()
    
    # Track the path taken
    path = []
    
    while open_list:
        # Get node with lowest heuristic value
        _, current = heapq.heappop(open_list)
        
        # Add current node to path
        path.append(current)
        
        # Check if the goal has been reached
        if current == goal:
            return path
        
        # Mark current node as visited
        visited.add(current)
        
        # Explore neighbors
        for neighbor, cost in graph[current]:
            if neighbor not in visited:
                heapq.heappush(open_list, (heuristic[neighbor], neighbor))
    
    # If the goal is not reachable, return None
    return None

# Run Best First Search
start_node = 'A'
goal_node = 'G'
path = best_first_search(start_node, goal_node)

# Output the path found
if path:
    print("Path found:", " -> ".join(path))
else:
    print("No path found.")
```
